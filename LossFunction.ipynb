```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.linspace(-5, 5, 100)

# loss function
y = 2 * np.sin(x) + np.cos(2*x)
```

```python
plt.plot(x, y)
plt.title('Loss Function');
```

```python
def dy(x):
    
    yPrime = 2 * np.cos(x) - 2 * np.sin(2 * x)
    
    return yPrime
```

```python
def gradientDescent(x_start, times, learningRate):
    lr = learningRate     
    x = x_start
    y = 2 * np.sin(x) + np.cos(2*x)
    
    for t in range(times):
        plt.plot(x, y, 'o')
        x = x - lr * dy(x)
        y = 2 * np.sin(x) + np.cos(2*x)
        
    return x
```

### x start from 0, do 100 times at learning rate = 0.03 

```python
plt.plot(x, y)
gradientDescent(0, 100, 0.03)
```

### Now randomly assign x in range -5 to 5 and find a local minimum of y. Do this again and again for 10 times, then we will probably find our global minimum of y.

```python
x = np.linspace(-5, 5, 100)
y = 2 * np.sin(x) + np.cos(2*x)
plt.plot(x, y)

local_Min = []

for test in range(10):
    x = np.random.randint(-5, 5)
    y = 2 * np.sin(x) + np.cos(2 * x)
    min_x = gradientDescent(x, 100, 0.03)
    min_y = 2 * np.sin(min_x) + np.cos(2 * min_x)
    
    local_Min.append([min_y, min_x])
    # place min_y before min_x for the next step: 
    # sorting the list by value of min_y

local_Min.sort()     
global_Min = list(reversed(local_Min[0]))
# reverse (y, x) to (x, y) 

print('When x = '+ str(global_Min[0]) +',\n' + 'we have the global minimum of y = ' + str(global_Min[1]) + '.')
```
